```python
import pandas as pd

# Load the dataset into a Pandas DataFrame
df = pd.read_csv('Housing.csv')
# Display the first few rows of the DataFrame
print(df.head())

# Get information about the DataFrame (column names, data types, non-null counts)
print(df.info())

# Get statistical summary of the numeric columns
print(df.describe())
# Handle missing values (drop or fill them)
df.dropna(inplace=True)  # Example: drop rows with missing values
# df.fillna(0, inplace=True)  # Example: fill missing values with 0

# Remove duplicates
df.drop_duplicates(inplace=True)

# Perform data transformations (e.g., change data types)
# df['column_name'] = df['column_name'].astype('int')  # Example: convert a column to integer type
import matplotlib.pyplot as plt
import seaborn as sns

# Example of creating a histogram
plt.hist(df['price'], bins=10)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Housing Prices')
plt.show()

# Example of creating a scatter plot
plt.hist(df['price'], bins=10)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Housing Prices')
plt.show()

# Use Seaborn for more advanced visualizations
sns.boxplot(x='bedrooms', y='price', data=df) # Example
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Boxplot of Price by Number of Bedrooms')
plt.show()


```
